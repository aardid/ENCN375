### To do: Simulate diferent flooding scenarios (think about on the assumptions given below). 

For New Zealand, where tidal ranges are substantial, it makes sense to model sea level rise plus the effects of tides. If you're projecting a 1-meter rise, also simulate 2-meter scenarios to capture the potential maximum sea level during high tides. This will give a more realistic assessment of future flooding risks and impacts on coastal infrastructure.
So, we will be modelling a 'High Tide Scenarios', or High Tides + Sea Level Rise. 

For example, if the tidal range is 1 meter, a 1-meter sea level rise could effectively result in water levels that are 2 meters higher during high tide. 

We are not considering storm surges or other extreme weather events that can push sea levels even higher during storms. A combination of sea level rise, high tides, and storm surges can create significant flooding risks.

- 'High Tide Scenarios', or High Tides + Sea Level Rise (+1m)
